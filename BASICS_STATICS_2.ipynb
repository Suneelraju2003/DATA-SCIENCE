1.	Load the patient dataset and show summary with info().

```python
import pandas as pd

# Load CSV files
patient_df = pd.read_csv("/content/Billing_Data.csv")
billing_df = pd.read_csv("/content/Patient_Data.csv")

# Show summary
print("Patient Dataset Info:")
patient_df.info()

print("\nBilling Dataset Info:")
billing_df.info()
```

2.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].

```python
billing_cols = ['PatientID', 'Department', 'Doctor', 'BillAmount']
patient_billing_df = billing_df[billing_cols].copy()

display(patient_billing_df)
```

3.	Drop administrative columns like ['ReceptionistID', 'CheckInTime'].

```python
cols_to_drop = ['ReceptionistID', 'CheckInTime']

# Drop only if columns exist
patient_df_clean = patient_df.drop(columns=[c for c in cols_to_drop if c in patient_df.columns])

display(patient_df_clean)
```

4.	Use groupby to find total bill amount per department.

```python
dept_bill = patient_billing_df.groupby('Department')['BillAmount'].sum()

print("Total Bill Amount per Department:")
display(dept_bill)
```

5.	Remove duplicate patient records based on PatientID.

```python
patient_unique = patient_df_clean.drop_duplicates(subset='PatientID')

display(patient_unique)
```

6.	Fill missing BillAmount values with the mean bill amount.

```python
patient_billing_df['BillAmount'] = pd.to_numeric(patient_billing_df['BillAmount'], errors='coerce')

mean_bill = patient_billing_df['BillAmount'].mean()

patient_billing_df.fillna({'BillAmount': mean_bill}, inplace=True)
display(patient_billing_df)
```

7.	Merge the billing dataset with patient dataset on PatientID.

```python
merged_df = pd.merge(patient_unique, billing_df, on='PatientID')
merged_df = merged_df.drop_duplicates(subset='PatientID')
billing_df.update(patient_billing_df)
print("Merged Dataset:")
display(merged_df)
```

8.	Concatenate an additional DataFrame that contains new patients for the current week (row-wise).

```python
new_patients_df = pd.DataFrame({'PatientID': [106,107], 'InsuranceCovered' : [1500, 1000], 'FinalAmount' : [5000, 4000] , 'Name' : ['Amul', 'Arun'], 'Department' : ['Cardiology', 'Orthopedics'], 'Doctor' : ['Dr. Lee', 'Dr. John'], 'BillAmount' : [4500, 3900],'ReceptionistID':[2, 3], 'CheckInTime' : ['2023-01-15 10:20', '2023-01-16 12:10']})

updated_patients = pd.concat([merged_df, new_patients_df], ignore_index=True)

display(updated_patients)
```

```python
final_df = updated_patients.drop(columns='FinalAmount')
display(final_df)
```

9.	Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).

```python
insurance_covered_list = (['Yes', 'Yes', 'Yes', 'Yes','Yes','Yes','Yes'])

additional_cols = pd.DataFrame({
    'InsuranceCovered': insurance_covered_list,
    'Final_Amount': final_df['BillAmount'] - final_df['InsuranceCovered']  # example calculation
})

# Column-wise concatenation
final_cleaned_df = pd.concat([final_df.reset_index(drop=True),
                      additional_cols.reset_index(drop=True)], axis=1)
```

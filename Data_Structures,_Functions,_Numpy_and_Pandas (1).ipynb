**11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:**

**- Use .info(), .describe()**

**- Select 'Name' and 'City' columns**

**- Drop 'City' column**

**- Fill any missing values in 'Salary' column with the mean**

**- Remove any duplicate rows**

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ravi", "Sita", "Aman", "John", "Rani",
             "Meera", "Asha", "Sara", "Kiran", "Rani"],
    "Age": [25, 30, 22, 28, 25, 27, 24, 30, 29, 25],
    "City": ["Delhi", "Mumbai", "Chennai", "Delhi", "Delhi",
             "Hyderabad", "Pune", "Mumbai", "Chennai", "Delhi"],
    "Salary": [50000, 60000, np.nan, 55000, 50000,
               58000, 52000, np.nan, 61000, 50000]
}

df = pd.DataFrame(data)


print("DataFrame Info:")
print(df.info())


print("\nStatistics:")
print(df.describe())


print("\nSelected Columns (Name, City):")
print(df[['Name', 'City']])


df_dropped = df.drop('City', axis=1)
print("\nDataFrame after dropping 'City':")
print(df_dropped)


mean_salary = df['Salary'].mean()
df['Salary'] = df['Salary'].fillna(mean_salary)

print("\nDataFrame after filling missing Salary values:")
print(df)


df_no_duplicates = df.drop_duplicates()
print("\nDataFrame after removing duplicates:")
print(df_no_duplicates)
```
